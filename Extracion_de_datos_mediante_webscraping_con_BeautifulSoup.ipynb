Realización de Web Scraping con Python por medio de la herramienta BeautifulSoup. 
Se realiza la extracción de datos a la página web de wikipedia sobre las compañías que hacen parte del índice bursátil COLCAP




```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la pagina de Wikipedia con los datos de la bolsa de valores
url = 'https://es.wikipedia.org/wiki/COLCAP'

# Realizar solicitud GET a la página web y obtener el HTML
response = requests.get(url)
html = response.text

# Analizar el HTML con BeautifulSoup
soup = BeautifulSoup(html, 'lxml')
soup
```

```python
#Busqueda y extraccion de tabla sobre las compañias por las que esta compuesto el indice COLCAP
table = soup.find('table',{'class':'wikitable sortable'}) 
print(table)
```

```python
#Creacion de los titulos de cada uno de las columnas
company_data = pd.DataFrame(columns=["Compañia", "Bolsa de Valores", "Simbolo", "Industria"])
```

```python
#Se recorre cada fila y se asignan los valores a cada columna.
for row in soup.find("tbody").find_all('tr')[1:]:
    col = row.find_all("td")
    company = col[0].text.replace("\n", "").replace("\t", "")
    stock_market = col[1].text.replace("\n", "").replace("\t", "")
    symbol = col[2].text.replace("\n", "").replace("\t", "")
    industry = col[3].text.replace("\n", "").replace("\t", "")

#Finalmente se agregan los datos a cada fila de la tabla
    company_data = company_data.append({"Compañia":company, "Bolsa de Valores":stock_market, "Simbolo":symbol, "Industria":industry}, ignore_index=True)    
```
